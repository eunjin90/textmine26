```python
import pandas as pd

# 1. 데이터 확인
# 파일 경로 설정
filename = '../data/bccard_201906.csv'
df = pd.read_csv(filename)
df.head(3)
```

```python
# 1. 앞부분 확인 (기본 5행)
print("--- 데이터 앞부분 (head) ---")
display(df.head(3))
```

```python
# 2. 뒷부분 확인 (기본 5행)
print("\n--- 데이터 뒷부분 (tail) ---")
display(df.tail(3))
```

```python
# 3. 데이터 수 확인
print(f"\n전체 데이터 행 수: {len(df)}개")
print(f"전체 데이터 열 수: {df.shape[1]}개")
```

```python
# 서울 거주 고객과 비거주 고객 분리
seoul_cust = df[df['CSTMR_MEGA_CTY_NM'] == '서울특별시']
non_seoul_cust = df[df['CSTMR_MEGA_CTY_NM'] != '서울특별시']

# 1. 서울 거주 고객 분석 테이블
seoul_table = pd.DataFrame({
    '구분': ['서울 거주 고객'],
    '고객수(데이터건수)': [len(seoul_cust)],
    '총 소비액(AMT)': [seoul_cust['AMT'].sum()],
    '총 카드이용건수(CNT)': [seoul_cust['CNT'].sum()]
})

# 2. 비거주 고객 분석 테이블
non_seoul_table = pd.DataFrame({
    '구분': ['비거주 고객'],
    '고객수(데이터건수)': [len(non_seoul_cust)],
    '총 소비액(AMT)': [non_seoul_cust['AMT'].sum()],
    '총 카드이용건수(CNT)': [non_seoul_cust['CNT'].sum()]
})

print("--- [표 1] 서울 거주 고객 분석 ---")
display(seoul_table)

print("\n--- [표 2] 비거주 고객 분석 ---")
display(non_seoul_table)
```

```python
# 데이터 전처리: 업종명의 공백 제거 ('편 의 점' -> '편의점')
df['TP_BUZ_NM'] = df['TP_BUZ_NM'].str.replace(' ', '')

# 1. 전체 편의점 총 소비액
total_cvs_amt = df[df['TP_BUZ_NM'] == '편의점']['AMT'].sum()
print(f"1. 전국 편의점 총 소비액: {total_cvs_amt:,}원")

# 2. 강남구 전체 총 소비액 (가맹점 기준)
total_gangnam_amt = df[df['CTY_RGN_NM'] == '강남구']['AMT'].sum()
print(f"2. 강남구 지역 총 소비액: {total_gangnam_amt:,}원")

# 3. 강남구 소재 편의점 총 소비액
gangnam_cvs = df[(df['CTY_RGN_NM'] == '강남구') & (df['TP_BUZ_NM'] == '편의점')]
gangnam_cvs_amt = gangnam_cvs['AMT'].sum()
print(f"3. 강남구 내 편의점 총 소비액: {gangnam_cvs_amt:,}원")

# 4. 강남구 편의점 소비액 중 [강남구 거주자] vs [비거주자] 비교
# 'CSTMR_CTY_RGN_NM' 컬럼이 고객의 거주 구 정보를 담고 있습니다.
gn_res_amt = gangnam_cvs[gangnam_cvs['CSTMR_CTY_RGN_NM'] == '강남구']['AMT'].sum()
gn_non_res_amt = gangnam_cvs[gangnam_cvs['CSTMR_CTY_RGN_NM'] != '강남구']['AMT'].sum()

# 거주자/비거주자 소비액 차이 계산
amt_diff = gn_non_res_amt - gn_res_amt

# 결과 테이블 생성
comparison_table = pd.DataFrame({
    '구분': ['강남구 거주자', '강남구 비거주자', '액수 차이(비거주/거주)'],
    '강남구 편의점 소비액': [gn_res_amt, gn_non_res_amt, amt_diff]
})

print("\n--- [강남구 편의점] 거주자 vs 비거주자 소비 비교 ---")
display(comparison_table)
```
