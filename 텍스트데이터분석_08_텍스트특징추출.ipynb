# 텍스트 특징 수출 (TF-IDF)

### 1. 샘플 코퍼스로 연습

```python
sample_corpus = [
    '자연어처리 강의를 시작하겠습니다.',
    '자연어처리는 재미있습니다.',
    '밥을 먹고 강의를 듣고 있습니다.',
    '이번 자연어처리 강의는 한국어 자연어처리입니다.'
]
```

```python
import sklearn
```

```python
from konlpy.tag import Okt
def my_tokenizer(text):
    return Okt().nouns(text)
```

```python
# TfidfVectorizer 객체 생성
# from sklearn.feature_extraction.text import TfidfVectorizer
# vectorizer = TfidfVectorizer(tokenizer=my_tokenizer)
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(tokenizer=my_tokenizer)
# 특징 집합과 관련 데이터 모델을 생성
vectorizer.fit(sample_corpus)
print(vectorizer.get_feature_names_out())
# 특징 벡터 추출
sample_dtm = vectorizer.transform(sample_corpus)
print(sample_dtm.toarray())
```

### 2. 다음 영화 리뷰로 연습

```python
import pandas as pd

# 1. 데이터 불러오기
df = pd.read_csv('./data/daum_movie_review.csv')

# 2. 말뭉치(Corpus) 만들기
# 리뷰가 들어있는 컬럼 'review'를 추출하고, 빈 칸은 제거(dropna())한 뒤 리스트로 만든다
daum_corpus = df['review'].dropna().tolist()

# 3. 토크나이저 설정 (명사 추출 + 단어 길이 + 불용어 제거)
okt = Okt()
def my_tokenizer(text):
    nouns = okt.nouns(str(text)) # 명사만 추출

    stop_words = ['영화', '정말', '진짜', '관람', '보고', '최고', '점', '내내', '생각', '가가', '가감', '힌츠펠']

# 사진의 로직: 단어 길이가 2자 이상이고 불용어에 없는 것만 추출
    words = [n for n in nouns if len(n) >= 2 and n not in stop_words]
    return words

# 4. CountVectorizer 객체 생성
# 토크나이저를 연결
vectorizer = sklearn.feature_extraction.text.CountVectorizer(
    tokenizer=my_tokenizer, # 어떤 도구로 자를 것인가
    stop_words=None,        # 이미 함수에서 지웠으므로 None 또는 리스트 지정
    max_features=50,      # 상위 50개 단어만 사용
    min_df=2,               # 최소 2번 이상 등장한 단어만
    max_df=0.5              # 50% 이상 문장에 등장하는 너무 흔한 단어 제외
)

# 5. 특징 집합과 관련 데이터 모델을 생성
vectorizer.fit(daum_corpus)

# 6. 특징 이름(단어들) 확인
print(vectorizer.get_feature_names_out())

# 7. 특징 벡터 추출
# .transform() 숫자로 바꿈
daum_dtm = vectorizer.transform(daum_corpus)

# 8. 특징 벡터 결과 확인
print(daum_dtm.toarray()[:10, :10])
```
