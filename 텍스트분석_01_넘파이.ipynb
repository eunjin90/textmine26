### 2. 넘파이 배열 (ndarray) 생성과 속성 확인

```python
a1 = np.array([1, 2, 3])
print(a1, a1.dtype, a1.ndim, a1.shape, a1.size)
a2 = np.array((1, 2, 3))
print(a2, type(a2), a2.dtype, a2.ndim, a2.shape, a2.size)
```

```python
a3 = np.array([[1, 2, 3], [4, 5, 6]])
print(a3)
print(type(a3), a3.dtype, a3.ndim, a3.shape, a3.size)
```

```python
a1 = np.array([1, 2, 3], dtype=float)
print(a1, type(a1), a1.dtype, a1.ndim, a1.shape, a1.size)
```

```python
a6 = np.array([1, 4, 5, '8'])
print(a6)
print(type(a6), a6.dtype, a6.ndim, a6.shape, a6.size)
```

```python
a6 = np.array([1, 4, 5, '8'], int)
print(a6)
print(type(a6), a6.dtype, a6.ndim, a6.shape, a6.size)
```

```python
test_list = [[1, 4, 5, 8], [1, 4, 5]]
np.array(test_list)
```

### 3. Indexing

```python
d3 = np.arange(12).reshape(2, 2, 3)  # 0 ~ 11
print(d3)
```

```python
d3[0] = 50
print(d3)
```

```python
d3[1,0] = 30
print(d3)
```

### 4. slicing

```python
d2 = np.arange(1,11).reshape(2,-1)
print(d2)
d2[:,:] = -1
print(d2)
```

### 5. Boolean Indexing

```python
data = np.array([4, 6, 7, 3, 2, 0])
print(data > 3)
print(data[data > 3])
```

### 6. 넘파이 배열의 정렬

```python
# [연습문제] 성적순으로 출력 (내림차순)
names = np.array(['John', 'Mike', 'Sarah', 'Kate', 'Samuel'])
scores = np.array([78, 95, 84, 98, 88])

scores_sorted_index = np.argsort(scores)[::-1]
print(scores_sorted_index)

# 팬싱인덱스적용
print(names[scores_sorted_index])
```
